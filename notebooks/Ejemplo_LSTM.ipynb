# Ejemplo de datos de entrada y prediccion

```python
import pandas as pd
import tensorflow as tf
import numpy as np
import sys
import warnings
warnings.filterwarnings('ignore')
sys.path.append('../')
from src.utils.LSTMPipeline import LSTMPipeline
```

## Ejemplo con Cartago

```python
datos_entrada_cartago = pd.read_csv('../data/raw/ultimos_9_registros_cartago.csv', index_col=0)
datos_entrada_cartago
```

```python
pipeline = LSTMPipeline.load('../src/models/CARTAGO/pipeline.pkl')
X_new = pipeline.transform_predict(datos_entrada_cartago)


modelo = tf.keras.models.load_model('../src/models/CARTAGO/LSTM-CARTAGO.keras')

y_pred_scaled = modelo.predict(X_new)
prediccion = pipeline.inverse_transform_y(y_pred_scaled)
print(f'Se estiman {int(np.round(prediccion)[0][0])} delitos según las condiciones dadas.')
```
